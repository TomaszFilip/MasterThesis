```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv('normal_yolo.csv')  # Replace with your actual file path

# Compute FPR
df['FPR'] = df['accident_count'] / df['total_count']
df['A'] = df['A'] / 10

# Plot FPR vs A
plt.figure(figsize=(10, 6))
plt.plot(df['A'], df['FPR'], label='yolo11m', color='blue', marker='o')

# Plot formatting
plt.xlabel('a')
plt.ylabel('FPR')
plt.title('Acceleration change FPR(a)')
plt.grid(True)

plt.tight_layout()

df = pd.read_csv('normal_maskrcnn.csv')  # Replace with your actual file path

# Compute FPR
df['FPR'] = df['accident_count'] / df['total_count']
df['A'] = df['A'] / 10

plt.plot(df['A'], df['FPR'], label='Mask R-CNN', color='darkorange', marker='x')
plt.legend()
# Save plot
plt.savefig('plots/acceleration_fpr.png', dpi=300)  # Saves the figure as 'fpr_plot.png'

# Show plot
plt.show()
```
